```python
#ex11
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import numpy as np

def cmap(values):
	cmap = cm.get_cmap('rainbow')
	normalize = matplotlib.colors.Normalize(vmin=min(values), vmax=max(values))
	color = [cmap(normalize(value)) for value in values]
	return color

df = pd.read_csv('./Assignment 5 Data Files/gsales.csv', parse_dates=['Year'])
# print(df.head())

bar = df[(df['Global_Sales']>2) & (df['Genre']=='Sports') & (df['Year'].dt.year.isin(range(2010, 2016)))][['Year', 'Global_Sales']]
bar2010 = bar[bar['Year'].dt.year==2010]['Global_Sales'].tolist()
bar2011 = bar[bar['Year'].dt.year==2011]['Global_Sales'].tolist()
bar2012 = bar[bar['Year'].dt.year==2012]['Global_Sales'].tolist()
bar2013 = bar[bar['Year'].dt.year==2013]['Global_Sales'].tolist()
bar2014 = bar[bar['Year'].dt.year==2014]['Global_Sales'].tolist()
bar2015 = bar[bar['Year'].dt.year==2015]['Global_Sales'].tolist()

print(bar.head())
# bar.to_csv('./bt/ex10.csv', index=None)
bar = df.groupby(bar['Year'].dt.year)['Global_Sales'].sum().reset_index(name='total global sale')

print(bar)
colormap = bar['total global sale']
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(15,15))
ax1.bar(bar['Year'], bar['total global sale'], color=cmap(colormap))
ax1.set_ylabel('total global sale')

bar20102015 = [bar2010, bar2011, bar2012, bar2013, bar2014, bar2015]
label = range(2010,2016)
ax2.boxplot(bar20102015, vert=True, patch_artist=True, labels=label)
ax2.set_ylabel('Global sale range in period 2010-2015')
ax2.set_xlabel('Year')
fig.suptitle('Global sale in period 2010-2015', fontsize=20, color='blue')
fig.show()
```

```python
#Ex12
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import numpy as np

df = pd.read_csv('./Assignment 5 Data Files/sales_data.csv')
df = df[['Region', 'Segment', 'Sales']]
df = df.groupby(['Region', 'Segment'])['Sales'].sum().reset_index(name='sum sale')
print(df)


dictSale = {}
regionUnique = df['Region'].unique()
segmentUnique = df['Segment'].unique()

for segments in segmentUnique:
    listSale = []
    for re in regionUnique:
        if df[(df['Region']==re) & (df['Segment']==segments)].empty:
            sumSale = 0
        else:
            sumSale = df[(df['Region']==re) & (df['Segment']==segments)].iloc[0,-1]

        listSale.append(sumSale)
    dictSale[segments] = listSale
    

plt.figure(figsize=(18,8))
stackedBar = pd.DataFrame(index = regionUnique, data = dictSale)
print(stackedBar)

stackedBar.plot(kind="bar", stacked=True)
plt.xlabel('Region', fontsize=10)
plt.ylabel('Sale', rotation=0, fontsize=10)
plt.title('Stacked bar chart for Sales in each Regions stacked by Segments', fontsize=20, color='blue')
plt.show()
```

```python
#Ex19
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.cm as cm
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib

df = pd.read_csv('./Assignment 5 Data Files/OfficeSupplies.csv', parse_dates=['OrderDate'])
print(df.head())

def cmap(values):
    cmap = cm.get_cmap('rainbow')
    normalize = matplotlib.colors.Normalize(vmin=min(values), vmax=max(values))
    color = [cmap(normalize(value)) for value in values]
    return color

#report1
regionUnit = df.groupby(['Region', 'Rep'])['Units'].apply(lambda x:x.sum()).reset_index()
regionUnit.columns = ['Region', 'Rep', 'sum units']
regionUnit = regionUnit.pivot(index='Rep', columns='Region', values='sum units').fillna(0)
# regionUnit.set_index('Region')
print('\ntotal sold units of each representative in each region:\n', regionUnit)

regionUnit.plot(kind="bar", stacked=True)
plt.title('Stacked bar chart for Rep in each Regions stacked', color = 'blue', fontsize=20)
plt.xlabel('Rep')
plt.ylabel('sum order')
plt.show()

#report2
statistic = df.groupby('Region')['Units'].agg(['min', 'max', 'mean']).reset_index()
# statistic.columns = ['Region', 'min units in region', 'max units in region']
print('\nmaximum of total units in each region:\n', statistic)
statistic.set_index('Region', inplace=True)

statistic.plot(kind='bar')
plt.title('Min max and mean in each region', color = 'blue', fontsize=20)
plt.xlabel('Region')
plt.ylabel('count order')
plt.show()

#report3
df['Revenue'] = df['Units']*df['Unit Price']
dateMonth = df.pivot(index='OrderDate', columns='Rep', values='Revenue').fillna(0)
print(dateMonth.head())

label = dateMonth.columns
plt.figure(figsize=(18,10))
plt.plot(dateMonth)
plt.xlabel('order date')
plt.ylabel('Revenue')
plt.title('Revenue on each Rep', fontsize=20, color = 'blue')
plt.legend(label, loc='upper right')
plt.show()

#report4
df = df[df['Item']!='Desk']#remove desk cause having unit price hight
itemPrice = df.groupby('Item')['Unit Price'].unique().reset_index(name='price list')
print('item with unit price', itemPrice)

label = itemPrice.iloc[:,0].tolist()
itemPrice = itemPrice.iloc[:,-1].tolist()
print( label)

plt.figure(figsize=(18,6))
plt.boxplot(itemPrice, vert=True, patch_artist=True, labels=label)
plt.title('item with unit price', color = 'blue', fontsize=20)
plt.xlabel('Items', fontsize=15)
plt.ylabel('Price', fontsize=15, rotation=0) 
plt.show()

#report5

np.random.seed(0)
colorList = np.random.randint(1,100,df['Rep'].nunique())
marker = ['.','o','v','^','<','>','1','2','3','4','s','p','*','h','H','+','x','D','d','|','_']
markerList = np.random.choice(marker, df['Rep'].nunique())

plt.figure(figsize=(15,7))   
for index, rep in enumerate(df['Rep'].unique()):
    plt.scatter(df[df['Rep']==rep]['Units'], df[df['Rep']==rep]['Unit Price'], c=cmap(colorList), marker=markerList[index], s=100)

# handles = [mpatches.Patch(color=cmap(colorList), label=df['Rep'].unique())]
plt.legend(df['Rep'].unique() ,loc='upper right')
plt.title('Units and Unit Price in each rep', color = 'blue', fontsize=20)
plt.xlabel('Units', fontsize=15)
plt.ylabel('Unit Price', rotation=0, fontsize=15)
plt.show()
```

```python
#Ex20
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib

df = pd.read_csv('./Assignment 5 Data Files/UK Bank Customers.csv', parse_dates=['Date Joined'])
# print(df.head())

def cmap(values):
    cmap = cm.get_cmap('rainbow')
    normalize = matplotlib.colors.Normalize(vmin=min(values), vmax=max(values))
    color = [cmap(normalize(value)) for value in values]
    return color

#report1
customerRegion = df.groupby(['Region'])['Gender'].value_counts().unstack().reset_index()
customerRegion.set_index('Region', inplace=True)
print('\nstatistic customer each region and gender:\n', customerRegion)

plt.figure(figsize=(20,5))
customerRegion.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('people')
plt.xticks(rotation=45)
plt.title('Statistic customer each region and gender', fontsize=20, color='blue')
plt.show()

#report2
jobRegion = df.groupby(['Region', 'Job Classification'])['Job Classification'].agg(['count']).reset_index()
jobRegion =  jobRegion.groupby('Region')['Job Classification', 'count'].apply(lambda x: x.sort_values('count', ascending=False)).reset_index()
jobRegion.columns = ['Region', 'level_1', 'Job Classification', 'total customer']
jobRegion = jobRegion.drop(['level_1'], axis=1)
jobRegion = jobRegion.pivot(index='Region', columns='Job Classification', values='total customer')
print('\nstatistic job classification of each region:\n', jobRegion)

plt.figure(figsize=(20,5))
jobRegion.plot(kind='bar', stacked=True)
plt.xlabel('Country')
plt.ylabel('people')
plt.xticks(rotation=45)
plt.title('Statistic customer each region and Job Classification', fontsize=20, color='blue')
plt.show()

#report3
customerMonth = df.groupby(df['Date Joined'].dt.month)['Customer ID'].count().reset_index()
customerMonth.columns=['Month', 'total Customer']
print('\nstatistic customer join date each month:\n', customerMonth)

plt.figure(figsize=(20,5))
plt.barh(customerMonth['Month'], customerMonth['total Customer'], color=cmap(customerMonth['Month']))
plt.xlabel('Total Customer', fontsize=15)
plt.ylabel('Month', fontsize=15, rotation=0)

plt.title('Statistic customer each month', fontsize=20, color='blue')
plt.show()

#report4
biggestBalance = df.groupby('Region')['Balance'].agg(['max', 'min']).reset_index()
biggestBalance['error'] = 0.5*(biggestBalance['max']-biggestBalance['min'])
biggestBalance['mean'] = 0.5*(biggestBalance['max']+biggestBalance['min'])
print('\nsort mean of balance in each region:\n', biggestBalance)

plt.figure(figsize=(18,6))
plt.barh(biggestBalance['Region'], biggestBalance['mean'], xerr= biggestBalance['error'], color=cmap(biggestBalance['mean']))
plt.xlabel('Mean', fontsize=15)
plt.ylabel('Region', fontsize=15, rotation=0)
plt.title('Statistic mean of balance in each region', fontsize=20, color='blue')
plt.show()

#report5

def cmap(values):
    cmap = cm.get_cmap('rainbow')
    normalize = matplotlib.colors.Normalize(vmin=min(values), vmax=max(values))
    color = [normalize(value) for value in values]
    return color

colorcode = df['Balance']
plt.scatter(df['Age'], df['Balance'], c=cmap(colorcode))
plt.axis([10,70,10,190000])
plt.xlabel('Age', fontsize=15)
plt.ylabel('Balance', fontsize=15)
plt.title('Balance in each age', fontsize=20, color='blue')
plt.colorbar()
plt.show()
```
